# 大手証券会社の「待遇指標」

待遇指標とは「売上総利益 ÷ 人員数 ÷ 2」により算出される指標である。社員一人当たりの粗利に注目した指標であり、社員の待遇を大まかに把握することができる。

```python
# 必要なライブラリをインポートする
!pip install japanize-matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
corps = ['野村HD', '大和証券G本社', '三菱UFJフィナンシャル・G', 'SMBC日興証券株式会社', 'みずほフィナンシャルG']
years = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
grosses = np.array([[1557066, 1604178, 1395685, 1403201, 1496971, 1116770, 1287829, 1401872, 1363890, 1335577],
                    [541951, 532220, 514815, 472750, 505350, 441240, 426259, 466660, 502093, 464226],
                    [3753400, 4229000, 4143200, 4011800, 3854200, 3725700, 3986300, 3920900, 3964000, 4503000],
                    [352607, 351757, 318791, 361232, 397405, 369525, 398749, 447155, 355123, 279492],
                    [2035200, 2247700, 2221600, 2092700, 1915300, 1812700, 2062200, 2198600, 2252400, 2278400]])
employees_num = np.array([[27670, 28672, 28865, 28186, 28048, 27864, 26629, 26402, 26585, 26775],
                          [13317, 13466, 13701, 13836, 14791, 15196, 15320, 15096, 14889, 14731],
                          [127122, 135049, 138161, 138570, 119390, 117321, 115275, 110936, 108153, 106141],
                          [8884, 9367, 9685, 10112, 11987, 11759, 11484, 11241, 11105, 10871],
                          [54911, 54784, 56375, 59179, 60051, 59132, 57264, 54492, 52420, 51212]])
benefits = grosses / employees_num / 2
df = pd.DataFrame(benefits, columns=years).T

df.plot(marker='.', markersize=10)
plt.title('待遇指標（一人当たり売上総利益 ÷ 2）')
plt.xlabel('(年)')
plt.ylabel('(百万円/人)')
plt.legend(corps)
plt.show()
```
